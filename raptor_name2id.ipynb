```python
import pandas as pd

raptor = pd.read_csv('17_18_raptor.csv')
player2id = pd.read_csv('NBA_Player_IDs.csv', encoding='unicode_escape')
```

```python
raptor.head(20)
```

```python
player2id.dropna(subset=['BBRefName'], inplace=True)

print(player2id[player2id['BBRefName'].str.contains('Joe Young')])
```

```python
raptor_ids = pd.Series(index=range(len(raptor)))

for idx, row in raptor.iterrows():
    player = row['PLAYER'].strip()
    matching_players = player2id[player2id['NBAName'] == player]
    if len(matching_players) != 1:
        print(idx, len(matching_players), player)
    else:
        raptor_ids[idx] = matching_players['BBRefID'].values[0]
        
```

```python
# Manually fill the rest

raptor_ids[71] = 'tuckepj01'  # PJ Tucker
raptor_ids[230] = 'jamesmi02'  # Mike James
raptor_ids[297] = 'morrima03'  # Marcus Morris
raptor_ids[447] = 'youngjo01'  # Joe Young
```

```python
raptor['PLAYER_ID'] = raptor_ids

raptor.to_csv('17_18_raptor_with_ids.csv', index=False)
```
